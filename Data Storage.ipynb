```python
import requests
import pandas as pd
import numpy as np
import json
import csv
import itertools
import sys
import config
from config import client_id
import time



#Symbols to retrieve in order to get data
symbols = ["SPY", "IVV", "/ES", "QQQ", "TQQQ", "SQQQ"]  
```

```python
def getQuotes(symbols_to_use = []):
    endpoint = r"https://api.tdameritrade.com/v1/marketdata/quotes"

    #define our payload
    payload =  {'apikey':client_id,
                'symbol':symbols}

    #make a request
    content = requests.get(url = endpoint, params = payload)
    return content.json()
```

```python
def dict_to_CSV(dw = {}):
    for key in dw:
        new_dict = dw[key]
        temp_keys = new_dict.keys()
    
    
    #have to change csv file name because it is \ES 
    if(key == "/ES"):
        csv_file = "data_ES.csv"
    else:
        csv_file = "data_{}.csv".format(key)
        
        
    with open(csv_file, 'w') as f:
        w = csv.DictWriter(f,temp_keys)
        w.writeheader()
        w.writerow(new_dict)
```

```python
#Had to Hard Code, change if there is change to symbols.

def make_dataFrames_for_Symbols():
    dict_for_keys = getQuotes(symbols_to_use = symbols)
    
    spy_dict = dict_for_keys["SPY"]
    spy_df = pd.DataFrame(spy_dict.values(), spy_dict.keys())
    
    ivv_dict = dict_for_keys["IVV"]
    ivv_df = pd.DataFrame(ivv_dict.values(), ivv_dict.keys())
    
    es_dict = dict_for_keys["/ES"]
    es_df = pd.DataFrame(es_dict.values(), es_dict.keys())
    
    qqq_dict = dict_for_keys["QQQ"]
    qqq_df = pd.DataFrame(qqq_dict.values(), qqq_dict.keys())
    
    tqqq_dict = dict_for_keys["TQQQ"]
    tqqq_df = pd.DataFrame(tqqq_dict.values(), tqqq_dict.keys())
    
    sqqq_dict = dict_for_keys["SQQQ"]
    sqqq_df = pd.DataFrame(sqqq_dict.values(), sqqq_dict.keys())
    
    return (spy_df, ivv_df, es_df, qqq_df, tqqq_df, sqqq_df)
        
        
```

```python

def multi_quote():
    spy_df, ivv_df, es_df, qqq_df, tqqq_df, sqqq_df = make_dataFrames_for_Symbols()
    
    #60 x minutes
    #23400 = time market is open from 6:30 to 1:00
    t_end = time.time()+ 23400;
    
    #ROW OF DF TO USE, CHANGE BEFORE RUNNING TO PREVENT OVERRIDE
    x = 0
    
    while time.time()< t_end:
        #Get Quote (dict of dict), get single symbol(dict), make df (pd df), add to existing df
        try:
            full_size_dict = getQuotes(symbols_to_use = symbols)
            spy_df[x] = full_size_dict['SPY'].values()

            ivv_df[x] = full_size_dict['IVV'].values()

            es_df[x] = full_size_dict['/ES'].values()

            qqq_df[x] = full_size_dict['QQQ'].values()

            tqqq_df[x] = full_size_dict['TQQQ'].values()

            sqqq_df[x] = full_size_dict['SQQQ'].values()
            
        except KeyError:
            print("key Error, waiting 1 minute before resuming")
            time.sleep(60)
        except:
            print("other error, waiting 1 minute before resuming")
            time.sleep(60)
        
        if x% 25 == 0:
            print("Saving")
            spy_df.to_csv("data_SPY", header = True)
            ivv_df.to_csv("data_IVV", header = True)
            es_df.to_csv("data_ES", header = True)
            qqq_df.to_csv("data_qqq", header = True)
            tqqq_df.to_csv("data_tqqq", header = True)
            sqqq_df.to_csv("data_sqqq", header = True)
        
        x +=1 
        
        time.sleep(1.5)
        
    return (spy_df, ivv_df, es_df, qqq_df, tqqq_df, sqqq_df)
```

```python

if __name__=="__main__":
    spy_df, ivv_df, es_df, qqq_df, tqqq_df, sqqq_df = multi_quote()
    
    spy_df.to_csv("data_SPY", header = True)
    ivv_df.to_csv("data_IVV", header = True)
    es_df.to_csv("data_ES", header = True)
    qqq_df.to_csv("data_qqq", header = True)
    tqqq_df.to_csv("data_tqqq", header = True)
    sqqq_df.to_csv("data_sqqq", header = True)
    
```
